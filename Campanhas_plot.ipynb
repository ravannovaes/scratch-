```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import cycle
import itertools
```

## Gráfico 1 

Gráfico de taxa com e sem CA contíguo (IP Tput vs Distância) para três taxa oferecidas (500 Mbps, 1GBps, 2 Gbps) - 27 GHz .

```python
dfputIp = pd.read_csv('/home/linuxlite/results_service_NreMBB_SimPerScriptCampaign_TxPower23_SH_CH_xValue/data/Tput-IP.csv',  delimiter=',')
dfputIp=dfputIp.transpose()
new_header = dfputIp.iloc[0] #grab the first row for the header
dfputIp = dfputIp[1:] #take the data less the header row
dfputIp.columns = new_header #set the header row as the df header
dfputIp.head(10)
```

```python
dfTputx= pd.DataFrame()
dfTputx= pd.concat([dfTputx,
                    dfputIp['x']],axis=1)
dfTputx.head(10)
```

```python
dfputIp =dfputIp.drop(columns= 'x')
```

```python
dfputIp.head(10)
```

```python
#fontSize = 10
#figsize=(10,5)
#plt.figure(figsize=figsize)
#plt.rc('font', size=fontSize)  

markersA = cycle(('o','v', 'D', '>', 'h', '^', '*', '>', 's', 'H', 'D', 'd'))
colors = cycle(('b', 'g', 'r', 'c', 'm', 'y', 'k'))
linestyle = cycle(('-', '--'))# '-.', ':'))


markers = cycle(markersA)
dfmetric = list(dfputIp)
color=next(colors)
linestyle_on =next(linestyle)
markers_on = next(markers)



for i in dfmetric :           
        plt.plot(dfTputx,dfputIp[i],color=color,ls=linestyle_on, marker=markers_on )
        color=next(colors)
        linestyle_on=next(linestyle)
       # labelplot2 = next(labelplot)
       #print(color)
       # print(dfTputIp2)
        markers_on = next(markers)
        plt.legend(["27GHz, sem CA, 500Mbps","27GHz, sem CA, 1Gbps","27GHz, sem CA, 2Gbps",
                    "27GHz, com CA, 500Mbps","27GHz, com CA, 1Gbps","27GHz, com CA, 2Gbps" ])
plt.tight_layout()
plt.xlabel("Distância entre Sta e GNb")
plt.ylabel("Throughput kbps")
plt.grid()
   
```
